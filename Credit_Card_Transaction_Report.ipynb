```python
#IMPORTING CSV                #credit_card
import pandas as pd
df = pd.read_csv('credit_card.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
#FINDING NULL VALUES
df.isnull().sum()
```

```python
#FORMATTNG COLUMNS NAME
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```
